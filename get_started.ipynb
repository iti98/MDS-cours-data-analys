```python
notes = [ 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness' ]
data[notes]
```

```python
data[notes].mean()
# Pas ce qu'on veut
```

```python
data["mean_note"] = data[notes].mean(axis=1)
```

```python
data.sort_values("mean_note")
```

```python
gb_mean_notes = data.groupby("satisfaction")["mean_note"].mean()

gb_mean_notes
```

```python
sns.barplot(x=gb_mean_notes.index, y=gb_mean_notes.values)
```
